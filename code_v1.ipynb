```python
import h5py

with h5py.File('pbmc3k.h5ad', 'r') as f:
    # Examine obs (observation metadata)
    print("=== OBS (Observations) ===")
    obs_group = f['obs']
    print("Keys in obs:", list(obs_group.keys()))
    
    # Look at each key in obs
    for key in obs_group.keys():
        print(f"\nobs['{key}']:")
        data = obs_group[key]
        print(f"  Shape: {data.shape}")
        print(f"  Dtype: {data.dtype}")
        if data.size < 20:  # If small, print all values
            print(f"  Values: {data[:]}")
        else:  # If large, print first few
            print(f"  First 10 values: {data[:10]}")
    
    # Examine var (variable metadata)
    print("\n=== VAR (Variables) ===")
    var_group = f['var']
    print("Keys in var:", list(var_group.keys()))
    
    for key in var_group.keys():
        print(f"\nvar['{key}']:")
        data = var_group[key]
        print(f"  Shape: {data.shape}")
        print(f"  Dtype: {data.dtype}")
        if data.size < 20:
            print(f"  Values: {data[:]}")
        else:
            print(f"  First 10 values: {data[:10]}")
    
    # Examine uns (unstructured metadata)
    print("\n=== UNS (Unstructured) ===")
    uns_group = f['uns']
    print("Keys in uns:", list(uns_group.keys()))
    
    for key in uns_group.keys():
        print(f"\nuns['{key}']:")
        data = uns_group[key]
        print(f"  Type: {type(data)}")
        if hasattr(data, 'shape'):
            print(f"  Shape: {data.shape}")
        if hasattr(data, 'dtype'):
            print(f"  Dtype: {data.dtype}")
        # Try to read the data
        try:
            content = data[:]
            if isinstance(content, bytes):
                content = content.decode('utf-8')
            print(f"  Content: {content}")
        except:
            print("  (Cannot easily display content)")
```
